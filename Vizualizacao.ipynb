```python
#import lib
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Configurando o estilo do gráfico
sns.set_style("darkgrid")
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

```python
# Dados
ano = range(2011, 2019)
messi = [59,91,45,58,52,59,54,51]
cr7 = [60,63,69,91,57,55,53,49]

# primeiro grafico de linha
plt.plot(ano, messi, 'b-x', linewidth = 4, markersize = 12, markeredgewidth = 4, markeredgecolor = 'navy')

#segundo grafico de linha
plt.plot(ano, cr7, 'r-o', linewidth = 4, markersize = 12)

#titulo
plt.title('Messi vc CR7')

#legenda
plt.legend(['Messi', 'CR7'])

#eixos
plt.xlabel('Ano')
plt.ylabel('Gols')

#Salvar a imagem
plt.savefig('grafico.png')
```

```python
import pandas as pd

iris = pd.read_csv('/content/Iris.csv')
```

```python
iris.head()
```

```python
#scatter plot
sns.scatterplot(data=iris, x="SepalLengthCm", y="SepalWidthCm", hue="Species")

plt.title("Flores")
```

```python
plt.title("Distribuição Sepal Width")

sns.displot(iris.SepalWidthCm, kde=False);
```

```python
#grafico de distribuição
swarm_plot = sns.distplot(iris.SepalWidthCm, kde = False)
fig = swarm_plot.get_figure()
fig.savefig('out.png')
```

```python
#pegando nome das funções de gráficos
sns.get_dataset_names()
```

```python
#mapa de calor
sns.heatmap()
```

```python
#carregar dados do voo
#reformule od dados (produza uma tabela "pivot") com base nos valores da coluna. Usa valores exclusivos do índice especificado
voos = sns.load_dataset("flights").pivot_table(index='month',
                                                columns='year',
                                                values='passengers')
#titulo
plt.title("Número de Passageiros (1000)")
#desenha o grafico
sns.heatmap(voos,
            fmt=".0f",
            annot=True,
            linewidths=.5,
            cmap='Reds',
            annot_kws={"fontsize":10})
```

```python
#contour plot
plt.title("Flores")

sns.kdeplot(data= iris, x='SepalLengthCm', y='SepalWidthCm', shade=True, shade_lowest=False)
```

```python
plt.title("Flores")

sns.kdeplot(data= iris, x='SepalLengthCm', y='SepalWidthCm', hue='Species', fill=True)
```
